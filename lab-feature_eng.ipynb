**Check for missing values**

```python
spaceship.isna()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
df_cleaned=spaceship.dropna()
df_cleaned
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
df_cleaned['Cabin'] = df_cleaned['Cabin'].str[0]
df_cleaned
```

- Drop PassengerId and Name

```python
df_cleaned.drop(columns=['PassengerId', 'Name'], inplace=False)
```

- For non-numerical columns, do dummies.

```python
# Convert non-numerical 'Cabin' column into dummy/indicator variables
df_dummies = pd.get_dummies(df_cleaned, columns=['Cabin','HomePlanet','Name'])
df_dummies
```

**Perform Train Test Split**

```python
df_num=df_dummies.select_dtypes(include=['number']) 
df_num
```

```python
features = df_num # This is the new table for the X values to compare against the Target
target = df_dummies["Transported"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
y_train.head()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train) # Kind of like a cookie cutter, cutting the data to the shape of the cookie
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train_norm, y_train)
```

```python
pred = knn.predict(X_test)
pred
```
